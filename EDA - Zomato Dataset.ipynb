```python
df_country= pd.read_excel('Country-Code.xlsx')
df_country
```

```python
final_df= pd.merge(df,df_country,on='Country Code',how='left')
final_df
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
Country_index= final_df.Country.value_counts().index
Country_index
```

```python
Country_value= final_df.Country.value_counts().values
Country_value
```

```python
#Drawing a pie chart to show country and their values
plt.pie(Country_value, labels= Country_index)
```

```python
# pie chart for top three countries that uses zomato
plt.pie(Country_value[:3], labels= Country_index[:3],autopct='%1.2f%%') # autopct='%1.2f%%' IS USED TO SHOW PERCENTAGE IN PIE-CHART
                                                                      # 1.2 IS USED TO DISPLAY DECIMAL UPTO 2 DECIMAL PLACES
```

#### Observation: Zomato maximum records or transaction are from India, after that USA and then United Kingdoms.

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings= final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating count"}) 
ratings
```

### Observation
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15,12) # This code handles the size of barplot
sns.barplot(x="Aggregate rating", y="Rating count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating count",hue="Rating color", data=ratings, palette=['blue','red','orange','yellow','green','green' ])
```

### Observation:

1. Not Rated count is very high
2. Maximum number of rating are between 2.7 to 4.1

```python
#Count Plot
sns.countplot(x="Rating color",data= ratings,palette=['blue','red','orange','yellow','green','green'])
```
